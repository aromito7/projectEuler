```python
output_70 = float('inf')
primes = []
def factorize(num):
    factors = []
    p = 2
    while p < num:
        if num % p == 0:
            factors += [p]
        p += 1
    return factors


def totient(num):
    output = num
    factors = []
    for i in range(2, num):
        if num % i == 0:
            factors += [i]
            while(num % i == 0):
                num //= i
    
    for f in factors:
        output *= (f-1)/f
    return int(output)

def is_permutation(n1, n2):
    if n1 == n2: return False
    n1, n2 = [s for s in str(n1)], [s for s in str(n2)]
    n1.sort()
    n2.sort()
    return n1 == n2
    

for n in range(2, 10**5):
    tot = totient(n)
    for prime in primes:
        if n % prime == 0:
            continue
        primes += [n]
    if(is_permutation(n, tot)):
        if n/tot < output_70:
            output_70 = n/tot
            print(n, tot, factorize(n))
        
#is_permutation(totient(87109), 87109)
```

```python
length = 10**7

prime = [True for i in range(length)]
primes = []
prime[0], prime[1] = False, False

for n in range(2, length):
    if not prime[n]: continue
    primes += [n]
    for i in range(n**2, length, n):
        prime[i] = False
print(primes)

def is_permutation(n1, n2):
    if n1 == n2: return False
    n1, n2 = [s for s in str(n1)], [s for s in str(n2)]
    n1.sort()
    n2.sort()
    return n1 == n2

minimum = float('inf')

for p1, prime1 in enumerate(primes):
    for p2, prime2 in enumerate(primes[p1:]):
        num = prime1 * prime2
        if num > length: break
        tot = (prime1-1) * (prime2-1)
        if is_permutation(num, tot) and num/tot < minimum:
            print(prime1, prime2, num, tot, num/tot)
            minimum = num/tot
```
