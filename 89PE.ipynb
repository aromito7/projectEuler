```python
answer_81 = 0
f = open("files/p089_roman.txt", "r")
nums = f.readlines()
nums = [str(num[:-2]) for num in nums]
values = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
roman_digits_required = {0: 0, 1: 1, 2: 2, 3: 3, 4: 2, 5: 1, 6: 2, 7: 3, 8:4, 9:2}

for num in nums:
    #print(num)
    total = 0
    digits = 0
    for i, n in enumerate(num):
        current = values[n]
        if i == len(num) - 1 or current >= values[num[i+1]]:
            total += current
        else:
            total -= current
    #print(total)
    t = total

    for i in range(3, -1, -1):
        temp = (total//(10**i))%10
        rds = roman_digits_required[temp]
        if i == 3:
            rds = temp
        digits += rds
    #print(digits, len(num))
    temp = len(num) - digits
    if temp: print(num, t, len(num), digits)
    answer_81 += temp

print(answer_81)
```
