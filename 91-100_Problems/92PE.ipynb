```python
answer_92 = 0

limit = 10**7
ending = {1 : 1, 89: 89}

def next(num):
    n = sum([int(i)**2 for i in str(num)])

    #print(num, n)
    if n in ending: return ending[n]  
    else:
        ending[n] = next(n)
        return ending[n]


for i in range(1, 0):
    #print(i)
    if next(i) == 89: answer_92 += 1

#print(answer_92)

for n1 in range(0, 10):
    for n2 in range(n1, 10):
        for n3 in range(n2, 10):
            for n4 in range(n3, 10):
                for n5 in range(n4, 10):
                    for n6 in range(n5, 10):
                        for n7 in range(n6, 10):
                            n = [n1, n2, n3, n4, n5, n6, n7]
                            n = [str(digit) for digit in n]
                            diff = [n[i] - n[i-1] for i in range(1, len(n))]
                            print(''.join(n), diff)

print(answer_92)
```
